### From temperature to phonon

#### Introduction

* In solids, heat carried by **phonons**. **Exception**: In **metals, electrons** also contribute significantly
* **Phonons** == quantized lattice vibrations propagating as waves
* **Temperature T** == measure of atoms' deviation from equilibrium as they vibrate


#### Heat capacity

* characterizing physical property
* w/ $\Delta Q$ thermal energy, $\Delta T$ change of temperature, $V$ volume: 
 $C_V = \frac{\Delta Q}{\Delta T}\Bigr|_{V=\text{const}}$
* **isochoric** == $(V=\text{const})$ 
* w/ $M$ mass: **specific heat capacity** $c_V = \frac{C_V}{M}$

#### Heat flow

##### Definition of heat flow

* **heat current** $J(c)$ flows from hot to cold in the form of **phonon flow**
* phonons move w/ average velocity $<\mathbf{v_x}>$ along $\mathbf{x}$
* phonons scatter at surfaces, defects, e/a other

##### Calculation of heat flow

We consider heat flow $\mathbf{J_x}$ through a plane $A$ at $x$:
![heat flow](img/l7_heatflow.png)

and observe:

* at e/a point $\mathbf{x}$, thermal energy $\mathbf{Q(x)}$ has 50% chance to move L or R
* net heat flow $\mathbf{J(x)}$ is sum of contributions from both sides:
$\begin{align}
\mathbf{J(x)} & = \mathbf{J_{\text{L}\rightarrow\text{R}}(x-l_x) + J_{\text{R}\rightarrow\text{L}}(x+l_x)} \\
              & = \frac{1}{2} \cdot ((Q(x - l_x) - Q(x + l_x)) \cdot \left< v_x \right> \cdot \frac{1}{V}
\end{align}$

then using:

* $Q(x - l_x) - Q(x + l_x) \approx -2 \cdot l_x \cdot \frac{dQ(x)}{dx}$
* $Q(x) = f(T(x))$ and hence $\frac{dQ(x)}{dx} = \frac{dQ}{dT}\frac{dT}{dx}$
* for $\tau$ avg time btw two scattering events: mean free path along $x$ is $l_x = \left< v_x \right>\cdot\tau$
* heat capacity $C_V = \frac{dQ}{dT}$

we get:
$\begin{align}
J(x) & = - l_x \frac{dQ}{dx} \left< v_x \right> \frac{1}{V} \nonumber \\
J(x) & = - l_x \frac{dQ}{dT}\frac{dT}{dx} \left< v_x \right> \frac{1}{V} \nonumber \\
J(x) & = - \left< v_x \right>^2 \tau C_V \frac{dT}{dx} \frac{1}{V} \nonumber \\
J(x) & = - \frac{C_V}{V} {\left< v_x \right>}^2 \tau \frac{dT}{dx} \nonumber \\
J(x) & = - \kappa\frac{dT}{dx}
\end{align}$

where we defined $\kappa$ as the thermal conductivity:

$ \kappa = \frac{C_V}{V}{\left< v_x \right>}^2 \tau$

or in 3D w/ $\left< v_x \right> = \left< v_y \right> = \left< v_z \right>$ and $\mathbf{\left< v \right>}^2 = \left< v_x \right>^2 = \left< v_y \right>^2 = \left< v_z \right>^2$ and $\left< v \right> \tau = l_{\text{mfp}}$:

$ \kappa = \frac{C_V}{3V}{\left< v \right>} l_{\text{mfp}}$

#### Thermal conductivity

* **thermal conductivity** as defined above indicates how well a material conducts heat
* definition only valid if considered length $L \gg l_{\text{mfp}}$ the mean free path length: Macroscopic temperature $T$ unambigously defined
* $L \approx l_{\text{mfp}}$ requires **phonon population distribution**
